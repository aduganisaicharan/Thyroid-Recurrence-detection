```python
df=pd.get_dummies(df,columns=object_columns)
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_en = LabelEncoder()
```

```python
df['Recurred']= label_en.fit_transform(df['Recurred'])
```

```python
df['age_log'] = np.log(df['Age']+1)
df = df.drop('Age',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
X = df.drop('Recurred',axis=1)
y=df['Recurred']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,stratify = y)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
clf_rpt = classification_report(y_test,y_pred)
print(clf_rpt
     )
```

```python
# Support vector Classifier
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train,y_train)
```

```python
svc_ypred = svc_model.predict(X_test)
acc_svc = accuracy_score(y_test,svc_ypred)
print(acc_svc)
```

```python
clf_svc = classification_report(y_test,svc_ypred)
print(clf_svc)
```
